```python
import numpy as np
import matplotlib.pyplot as plt

A56, L56, B56, C56, M56 , N56 = 2, 3, 1, 0, 5, 6

n56 = np.linspace(0,100,101)

if N56 %2 != 0: #if odd
    #case:1 - 0<a<1 (decaying)
    exp_56 = 0.9**n56
    noise56 = np.random.rand(len(n56)) #uniform distribution noise
else: #if even
    #case:2 - a>1 (rising)
    exp_56 = 1.1**n56
    noise56 = np.random.randn(len(n56)) #normal distribution noise

    
noisy_signal_56 = exp_56 + 0.20 * exp_56 * noise56  # 20% noise

   
plt.figure(figsize=(16,10))
plt.subplot(2,2,1)
plt.plot(n56,noise56,'gray')
plt.title("Noise (ET231056)")
plt.xlabel("Time")
plt.ylabel("Amplitude")

plt.subplot(2,2,2)
plt.hist(noise56, bins=20, edgecolor='black', color='gray')
plt.title("Noise Distribution (ET231056)")
plt.xlabel("Time")
plt.ylabel("Amplitude")

plt.subplot(2,2,3)
plt.plot(n56,exp_56)
plt.title("Exponential Signal (ET231056)")
plt.xlabel("Time")
plt.ylabel("Amplitude")

plt.subplot(2,2,4)
plt.plot(n56,noisy_signal_56)
plt.title("Noisy Exponential Signal (ET231056)")
plt.xlabel("Time")
plt.ylabel("Amplitude")

plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

# Load your image (update the path to your image file)
image_path = r"Wazed.jpeg"  # Replace with your actual image path
image = cv2.imread(image_path)
# Convert BGR (OpenCV default) to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
# Generate Gaussian noise
mean = 0
std_dev = 25
noise = np.random.normal(mean, std_dev, image_rgb.shape)
# Create noisy image and clip values to valid range
noisy_image = np.clip(image_rgb + noise, 0, 255).astype(np.uint8)
# Visualize noise separately (scaled for display)
noise_display = np.clip(noise + 128, 0, 255).astype(np.uint8)
# Plotting
plt.figure(figsize=(15, 5))

# Original Image
plt.subplot(1, 3, 1)
plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis("off")

# Noise Image
plt.subplot(1, 3, 2)
plt.imshow(noise_display)
plt.title("Noise Image")
plt.axis("off")

# Noisy Image
plt.subplot(1, 3, 3)
plt.imshow(noisy_image)
plt.title("Noisy Image")
plt.axis("off")

plt.tight_layout()
plt.show()
```
