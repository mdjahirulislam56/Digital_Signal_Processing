```python
import matplotlib.pyplot as plt
import numpy as np

A,L,B,C,M,N=2,3,1,0,5,6

f = (M+N)
Fz = (M*5+N*3+L)
T = 1/Fz
duration = (M/2+N/4)
amplitude = (L+M+N)/5

t = np.arange(0,duration+T,T)
y = amplitude*np.sin(2*np.pi*f*t)

plt.figure(figsize=(16,12))
plt.subplot(2,2,1)
plt.plot(t,y,label='Sine Wave')
plt.title("Continuous plot of Sine Wave")
plt.xlabel("Time (t)") 
plt.ylabel("Amplitude") 
plt.grid(True)
plt.legend(loc='upper right')


F1= (L+N)
F2=(L+M) 
Fz2 = (L*3+N*2+M*2) 
T2 = 1/Fz2
duration2 = ((M+N)/4)

t2 = np.arange(0,duration2+T2,T2)
y2 = np.sin(2*np.pi*F1*t2) + np.sin(2*np.pi*F2*t2) 

plt.subplot(2,2,2)
plt.stem(t2,y2,linefmt="-",markerfmt='o',label='sin(2*pi*F1*t) + sin(2*pi*F2*t)')
plt.title("discrete plot of y= sin(2*pi*F1*t) + sin(2*pi*F2*t)")
plt.xlabel("Sample Position") 
plt.ylabel("Amplitude") 
plt.grid(True)
plt.legend(loc='upper right')

F3 = (A+B+C)
Fz3 = (L*3+N*4)
T3 = 1/Fz3
duration3 = ((M+N)/4)
amplitude3 = (L+M+N)/5

t3 = np.arange(0,duration3+T3,T3)
y3 = amplitude3*np.cos(2*np.pi*F3*t3)  

plt.subplot(2,2,3)
plt.stem(t3,y3,linefmt="r-",markerfmt='.-',basefmt='r-',label='Discrete Cosine Wave')
plt.plot(t3,y3,label='Continuous Cosine Wave')
plt.title("Continuous  And  Discrete  plot  of  Cosine  Wave")
plt.xlabel("Sample Position") 
plt.ylabel("Amplitude") 
plt.grid(True)
plt.legend(loc='upper right')


F4 = (A+B+C)
Fz4 = (L*3+N*4)
T4 = 1/Fz4
duration4 = (M+N)
amplitude4 = (L+M+N)/5

t4 = np.arange(-duration4,duration4+T4,T4)
y4 = amplitude4*np.sinc(F4*t4)

plt.subplot(2,2,4)
plt.stem(t4,y4,markerfmt='.',basefmt='r',label='Sinc Wave')
plt.title("Discrete  Plot  of  Sinc  Function")
plt.xlabel("Sample Position") 
plt.ylabel("Amplitude") 
plt.grid(True)
plt.legend(loc='upper right')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

A,L,B,C,M,N=2,3,1,0,5,6

x = np.arange(-(L+M),(L+N)+1)
y = np.where(x == 0, 1 , 0)

plt.figure(figsize=(16,15))
plt.subplot(3,1,1)
plt.stem(x,y,linefmt='-',markerfmt='o')
plt.title("Unit Sample Sequence")
plt.xlabel("Sample Position")
plt.ylabel("Amplitude")

x2 = np.arange(-(L+M),(L+N)+1)
y2 = np.where(x < 0, 0 , 1)

plt.subplot(3,1,2)
plt.stem(x2,y2,linefmt='-',markerfmt='o')
plt.title("Unit step sequence")
plt.xlabel("Sample Position")
plt.ylabel("Amplitude")

x3 = np.arange(0, (L+M+N)+1)
y3 = x3

plt.subplot(3,1,3)
plt.stem(x3,y3,linefmt='-',markerfmt='o')
plt.title("Unit ramp sequence")
plt.xlabel("Sample Position")
plt.ylabel("Amplitude")

plt.show()
```
