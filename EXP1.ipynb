```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,1,500)
x
```

```python
y = 3*x  
```

```python
plt.plot(x,y,label='y=mx')
plt.title('Exp1 practice')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.grid(True)
plt.legend(loc='upper right')
plt.show()
```

```python
y1 = np.sin(2*np.pi*5*x)

plt.plot(x,y1,label='sine wave')
plt.title('Exp1 practice 2')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.grid(True)
plt.legend(loc='upper right')
plt.show()
```

```python
plt.figure(figsize=(10,7))
plt.subplot(1,2,1)
plt.plot(x,y,label='y=mx')
plt.title('Exp1 practice')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.grid(True)
plt.legend(loc='upper right')

plt.subplot(1,2,2)
plt.plot(x,y1,label='sine wave')
plt.title('Exp1 practice 2')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.grid(True)
plt.legend(loc='upper right')
plt.show()
```
