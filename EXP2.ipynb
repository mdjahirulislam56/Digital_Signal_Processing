```python
import numpy as np
import matplotlib.pyplot as plt


initial_time = 0

Fs = 100 #sampling frequency
T = 1/Fs #Sampling period
F=2 #Frequency

duration = 1

final_time = duration+T

t = np.arange(initial_time,final_time,T)

Amplitude = 1

y = Amplitude*np.sin(2*np.pi*F*t) #Amplitude=1, Frequency=F, time=t

plt.stem(t,y,linefmt='-',markerfmt='o',label='stem')

plt.plot(t,y,linewidth=1,label='plot')

plt.legend()

plt.show()
```

```python
#sinc with sine function
import numpy as np
import matplotlib.pyplot as plt


initial_time = 0

Fs = 100 #sampling frequency
T = 1/Fs #Sampling period
F=2 #Frequency

duration = 1

no_of_steps = duration*Fs

t = np.linspace(-duration,duration,no_of_steps)

#y = np.sin(2*np.pi*F*t) / (2*np.pi*F*t)
#y[np.isnan(y)] = 1 #at sin(0) it will replace with 1 to avoid division by zero error

y = np.where(t == 0, 1, np.sin(2*np.pi*F*t)/(2*np.pi*F*t)) 

plt.stem(t,y,linefmt='-',markerfmt='o',label='stem')

plt.plot(t,y,linewidth=1,label='plot')

plt.legend()

plt.show()
```

```python
#using np.sinc 
initial_time = 0

Fs = 100 #sampling frequency
T = 1/Fs #Sampling period
F=2 #Frequency

duration = 1

no_of_steps = duration*Fs

t = np.linspace(-duration,duration,no_of_steps)

y = np.sinc(2*F*t)

plt.stem(t,y,linefmt='-',markerfmt='o',label='stem')

plt.plot(t,y,linewidth=1,label='plot')

plt.legend()

plt.show()
```

```python
#Exponential Signal;

#n = np.arange(0,101)
n = np.linspace(0,100,101)

#case:1 - 0<a<1
y1 = 0.9**n

#case:2 - a>1
y2 = 1.1**n

#case:3 - -1<a<0
y3 = (-0.9)**n

#case:4 - a<-1
y4 = (-1.1)**n

plt.figure(figsize=(10,8))
plt.subplot(2,2,1)
plt.title("case-1: - 0<a<1")
plt.stem(n,y1,linefmt='r-',markerfmt='r.',basefmt='b-') #case:1 plot

plt.subplot(2,2,2)
plt.title("case-2: - a>1")
plt.stem(n,y2,linefmt='b-',markerfmt='b.',basefmt='r-') #case:2 plot

plt.subplot(2,2,3)
plt.title("case-3: -1<a<0")
plt.stem(n,y3,linefmt='g-',markerfmt='g.',basefmt='k-') #case:3 plot

plt.subplot(2,2,4)
plt.title("case-4: - a<-1")
plt.stem(n,y4,linefmt='m-',markerfmt='m.',basefmt='k-') #case:4 plot

plt.show()
```
